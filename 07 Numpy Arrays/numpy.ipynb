## Arrays instead of list for mathematical calculations

### Multiplying by 3 t oa list vs array

```python
e = [12, 13, 14, 15, 16]
type(e)
```

```python
# in list  * operator means repeat list
e * 3
```

```python
e1 = []
for i in e:
    print(f"Multiplying {i} by 3 = {i * 3}")
    e1.append(i * 3)
```

```python
e1
```

```python
e2 = np.array(e)
e2
```

```python
type(e2)
```

```python
e3 = e2 * 3
print(e3)
```

```python
e2
```

```python
e2 + 5
```

```python
3 * e2 + 5
```

```python
h1
```

```python
h1.ndim
```

```python
h1.shape
```

```python
h1.sum()
```

```python
h1.mean()
```

```python
h1.std()
```

```python
h1.max()
```

```python
h1.min()
```

```python
h1.sum(axis=0)  # Sum along the first axis (columns)
```

```python
h1.mean(axis=0)  # Mean along the first axis (columns)
```

```python
h1.std(axis=0)  # Standard deviation along the first axis (columns)
```

```python
h1.min(axis=0)  # Minimum along the first axis (columns)
```

```python
h1.max(axis=0)  # Maximum along the first axis (columns)
```

```python
h1.prod()
```

```python
h1.prod(0)
```

```python
h1.prod(axis=1)  # Product along the second axis (rows)
```

```python
h1.mean(axis=1)  # Mean along the second axis (rows)
```

```python
h1.std(axis=1)  # Standard deviation along the second axis (rows)
```

```python
h1.max(axis=1)  # Maximum along the second axis (rows	)
```

```python
h1.min(axis=1)  # Minimum along the second axis (rows)
```

```python
h1.argmax(axis=1)  # Index of maximum along the second axis (rows)
```

```python
h1.argmax(axis=0)
```

```python
h1.argmin(axis=0)  # Index of minimum along the second axis (rows)
```

```python
h1.argmin(axis=1)  # Index of minimum along the second axis (rows)
```
